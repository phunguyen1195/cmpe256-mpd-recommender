```python
for path in glob.glob('data_withmeta/*.json'):
    basename = Path(path).stem
    track_meta = []
    print (path)
    f = open(path,mode='r')
    reader = jsonlines.Reader(f)
    for obj in reader.iter(type=dict, skip_invalid=True, allow_none=True):
        if obj is not None:
            track_meta.append (obj)
    name_check = 'data_smaller/' + basename.replace("_with_meta","") + '.csv'
    print (name_check)
    df = pd.read_csv(name_check)
    df_track_meta = pd.DataFrame(track_meta)
    df_track_meta_for_merge = df_track_meta.drop(columns=['id','track_href','analysis_url','duration_ms'])
    df_track_meta_for_merge = df_track_meta_for_merge.rename(columns={'uri':'track_uri'})
    df_track_merge_final = pd.merge(df,df_track_meta_for_merge,on='track_uri')
    df_track_merge_final.to_csv("data_withmeta/" +basename + ".csv", index=None)
```
